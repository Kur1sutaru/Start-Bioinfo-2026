## 1. Variáveis — preparando a caneca

Uma **variável** é como um rótulo colado num pote: você dá um nome e guarda um valor dentro.

```
nome_da_variavel = valor
```

````python
# Variáveis da nossa caneca
dono          = "Cristal"      # string  → texto entre aspas - mudem o nome aqui! rs
capacidade    = 300            # int     → número inteiro (ml)
nivel_atual   = 0              # int     → começa vazia
temperatura   = 90.5           # float   → número com casas decimais (°C)
caneca_cheia  = False          # bool    → True ou False

print(dono)
print(capacidade)
print(nivel_atual)
print(temperatura)
print(caneca_cheia)
````

---
## 2. Strings — descrevendo o café

**String** é qualquer sequência de texto. Você pode criar mensagens dinâmicas com **f-strings**:
coloque um `f` antes das aspas e use `{}` para inserir variáveis dentro do texto.

````python
# f-string: monta uma frase usando as variáveis
mensagem = f"A caneca de {dono} tem capacidade de {capacidade} ml e está com {nivel_atual} ml agora."
print(mensagem)

# Operações úteis com strings
cafe = "espresso"
print(cafe.upper())           # tudo maiúsculo
print(cafe.capitalize())      # primeira letra maiúscula
print(len(cafe))              # quantos caracteres tem
print(cafe.replace("e", "3")) # substituir letras (leetspeak ☕)
print(f"Tipo do café: {cafe!r}")  # !r mostra as aspas — útil pra depurar
````

---
## 3. Condicionais — verificar se precisa encher

**if / elif / else** permitem tomar decisões no código:
- `if` → se a condição for verdadeira, faz isso
- `elif` → senão, se outra condição for verdadeira, faz aquilo
- `else` → senão (nenhuma condição anterior), faz isso

````python
nivel_atual = 80  # ml na caneca agora

if nivel_atual == 0:
    print("🫙 Caneca vazia! Hora de encher.")
elif nivel_atual < capacidade * 0.25:   # menos de 25% da capacidade
    print("⚠️  Quase acabando o café...")
elif nivel_atual < capacidade:
    print(f"☕ Tem {nivel_atual} ml. Ainda dá pra tomar mais um gole.")
else:
    print(f"✅ Caneca cheia com {nivel_atual} ml!")
````

---
## 4. Loop while — tomando café até acabar

O **while** repete um bloco de código **enquanto** uma condição for verdadeira.
Cuidado: se a condição nunca mudar, o loop nunca para! (loop infinito 😱)

```python
while condição:
    # faz algo
    # atualiza a condição em algum momento!
```

````python
nivel_atual  = 300   # começa cheia
gole         = 50    # ml por gole
goles_dados  = 0

print(f"☕ Caneca cheia com {nivel_atual} ml. Vamos tomar café!\n")

while nivel_atual > 0:
    nivel_atual -= gole   # equivale a: nivel_atual = nivel_atual - gole
    goles_dados += 1

    # max(0, ...) garante que não fique negativo
    nivel_atual = max(0, nivel_atual)

    print(f"Gole {goles_dados}: tomou {gole} ml → restam {nivel_atual} ml")

print(f"\n😔 Caneca vazia depois de {goles_dados} goles!")
````

## 5. Funções — automatizando a rotina do café

Uma **função** é um bloco de código com nome, que você pode chamar várias vezes.
Isso evita repetição e deixa o código organizado.

```python
def nome_da_funcao(parametros):
    # corpo da função
    return resultado
```

````python
def verificar_caneca(nivel, capacidade):
    """Verifica o status da caneca e retorna uma mensagem."""
    porcentagem = (nivel / capacidade) * 100
    if nivel == 0:
        status = "vazia"
        emoji  = "🫙"
    elif porcentagem < 25:
        status = "quase vazia"
        emoji  = "⚠️"
    elif porcentagem < 75:
        status = "pela metade"
        emoji  = "☕"
    else:
        status = "cheia"
        emoji  = "✅"
    return f"{emoji} Caneca {status}: {nivel}/{capacidade} ml ({porcentagem:.0f}%)"


def encher_caneca(nivel, capacidade):
    """Enche a caneca até a capacidade máxima."""
    if nivel >= capacidade:
        print("A caneca já está cheia!")
        return nivel
    print(f"Enchendo: {nivel} ml → {capacidade} ml")
    return capacidade


def tomar_gole(nivel, gole=50):
    """Toma um gole. Parâmetro gole tem valor padrão de 50 ml."""
    if nivel == 0:
        print("Caneca vazia! Não tem café pra tomar. 😢")
        return 0
    novo_nivel = max(0, nivel - gole)
    print(f"Gole tomado! {nivel} ml → {novo_nivel} ml")
    return novo_nivel


# Testando as funções
capacidade = 300
nivel      = 0

print(verificar_caneca(nivel, capacidade))
nivel = encher_caneca(nivel, capacidade)
print(verificar_caneca(nivel, capacidade))
nivel = tomar_gole(nivel)
nivel = tomar_gole(nivel, gole=100)  # gole maior!
print(verificar_caneca(nivel, capacidade))
````

## 6. Listas — histórico de cafés do dia

Uma **lista** guarda vários valores numa ordem.
Você pode adicionar, remover e percorrer os itens.

````python
historico = []   # lista vazia

historico.append("07:30 — primeiro café ☕")
historico.append("10:00 — segundo café ☕")
historico.append("13:30 — café pós-almoço ☕")
historico.append("16:00 — último café (juro) ☕")

print(f"Cafés hoje: {len(historico)}\n")

for i, registro in enumerate(historico):
    print(f"  {i+1}. {registro}")

print(f"\nPrimeiro café do dia: {historico[0]}")
print(f"Último café do dia:   {historico[-1]}")
````

---
## 7. Juntando tudo — simulação completa da rotina

Agora vamos usar **tudo junto**: variáveis, strings, condicionais, loop e funções.

````python
# --- Configuração inicial ---
dono        = "Cristal"   #podemos mudar o nome aqui
capacidade  = 300
nivel       = 0
gole        = 60
historico   = []
hora        = 7

print(f"=== Rotina de café de {dono} ===")
print(f"Capacidade da caneca: {capacidade} ml | Gole: {gole} ml\n")

# --- Loop da rotina ---
while hora <= 18:

    # Enche se estiver vazia
    if nivel == 0:
        nivel = capacidade
        registro = f"{hora:02d}:00 — ☕ Encheu a caneca ({capacidade} ml)"
        historico.append(registro)
        print(registro)

    # Toma um gole
    nivel = max(0, nivel - gole)
    registro = f"{hora:02d}:00 — 🥤 Tomou um gole → restam {nivel} ml"
    historico.append(registro)
    print(registro)

    hora += 1  # avança 1 hora

# --- Resumo do dia ---
enchidas = sum(1 for r in historico if "Encheu" in r)
goles    = sum(1 for r in historico if "Tomou" in r)

print(f"""
=== Resumo do dia ===
Canecas enchidas : {enchidas}
Goles tomados    : {goles}
Total de café    : {enchidas * capacidade} ml
""")
````
