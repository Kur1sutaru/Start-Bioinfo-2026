## 🔧 Funções auxiliares (já prontas para você usar!)

```python
# Execute esta célula — define as funções que você pode usar

def parsear_fasta(fasta_string):
    """Faz o parse de uma string FASTA. Retorna dict {header: sequencia}."""
    sequencias = {}
    header_atual = None
    seq_atual = []
    for linha in fasta_string.strip().split('\n'):
        linha = linha.strip()
        if linha.startswith('>'):
            if header_atual:
                sequencias[header_atual] = ''.join(seq_atual)
            header_atual = linha[1:]
            seq_atual = []
        elif linha:
            seq_atual.append(linha)
    if header_atual:
        sequencias[header_atual] = ''.join(seq_atual)
    return sequencias


def calcular_gc(seq):
    """Calcula o GC content de uma sequência."""
    seq = seq.upper()
    return (seq.count('G') + seq.count('C')) / len(seq)


def complemento_reverso(seq):
    """Retorna o complemento reverso de uma sequência de DNA."""
    mapa = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    seq = seq.upper()
    comp = ''.join(mapa.get(b, 'N') for b in seq)
    return comp[::-1]


print('Funções carregadas com sucesso!')
```
